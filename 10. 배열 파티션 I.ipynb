### 561. Array Partition I
---
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi) for all i is maximized. Return the maximized sum.

Example 1:

Input: nums = [1,4,3,2]<br>
Output: 4<br>
Explanation: All possible pairings (ignoring the ordering of elements) are:<br>
> 1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3<br>
> 2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3<br>
> 3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4<br>
So the maximum possible sum is 4.<br>


Example 2:

Input: nums = [6,2,6,5,1,2]<br>
Output: 9<br>
Explanation: The optimal pairing is (2, 1), (2, 5), (6, 6). min(2, 1) + min(2, 5) + min(6, 6) = 1 + 2 + 6 = 9.

### 배열 파티션 I 
---
n개의 페어를 이용한 min(a, b)의 합으로 만들 수 있는 가장 큰 수를 출력하라

* __입력__
>  [1,4,3,2]
* __출력__
> 4
* __설명__
 n은 2개 되며, 최대 합은 4이다.
     min(1, 2) + min(3, 4) = 4

###  풀이 1. 오름차순 풀이 
---
문제가 상당히 특이하고 바로 이해하기가 쉽지 않다. 그러나 찬찬히 살펴보면 페어의 min()을 합산했을 때 최대(Argmax)를 만드는 것은 결국 min()이 되도록 커야한다는 뜻이고, 뒤에서부터 내림차순(Descending Order)으로 집어넣으면 항상 최대 min() 페어를 유지할 수 있다. 이 문제에서 배열 입력값은 항상 2n개일 것이므로 앞에서 부터 오름차순(Ascending Order)으로 집어넣어도 결과는 같을 것이다.


* __그림7-10__ min(a, b)의 합으로 최댓값을 만들 수 있는 조합

실제로 그림 7-10에서 하나씩 조합해보면 동일한 결과를 얻을 수 있다. 정렬된 상태에서 앞에서부터 오름차순으로 인접 요소 페어(Adjacent Elements Pair)를 만들면 가장 큰 a1과 a2페어들을 만들 수 있고 이 페어들의 합이 곧 만들 수 있는 최대 합이된다. 실제로도 그런지, 이 그림에서 만들어본 총 3가지 조합을 직접 계산해보면 다음과 같다.
    
    1. min(1,4) + min(2,3) = 3
    2. min(1,3) + min(2,4) = 3
    3. min(1,2) + min(3,4) = 4

여기서 세 번째가 가장 큰 값이 되며, 앞서 설명한 대로 정렬된 상태에서 앞에서부터 오름차순으로 페어를 만드면 된다. 뒤에서부터 내림차순으로 만들어도 결과는 동일하며, 여기서는 오름차순으로 구현해본다. 전체 코드는 다음과 같다.

```python
def arrayPairSum(self, nums: List[int]) -> int:
    sum = 0
    pair = []
    nums.sort()
    
    for n in nums:
        # 앞에서부터 오름차순으로 페어를 만들어서 합 계산
        pair.append(n)
        if len(pair) == 2:
            sum += min(pair)
            pair = []
    return sum 
```

### 풀이 2. 짝수 번째 값 계산
---
이렇게 페어에 대해 일일이 min() 값을 구하지 않아도 짝수 번째 값(0부터 시작하므로)을 더하면 될 것 같다. 정렬된 상태에서는 짝수 번째에 항상 작은 값이 위치하기 때문이다. 불필요한 리스트 변수를 생략할 수 있기 때문에 , 전체 코드 또한 많이 줄어들어 매우 간결하게 구현할 수 있다.

```python
def arrayPairSum(self, nums: List[int]) -> int:
    sum = 0 
    pair = []
    nums.sort()
    
    for i, n in enumerate(nums):
        # 짝수 번째 값의 합 계산
        if i % 2 ==0:
            sum += n
    return sum 
        
```
